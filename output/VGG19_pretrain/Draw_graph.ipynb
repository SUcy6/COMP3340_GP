```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Define the learning rate
lrs = [0.0001, 0.001, 0.01]
# Define the batch size
batch_sizes = [4, 32, 64]
```

```python
# read the data with lr = 0.0001
train_losses1 = pd.read_csv('./train_loss' + str(lrs[0]) + '.csv')
valid_losses1 = pd.read_csv('./valid_loss' + str(lrs[0]) + '.csv')
train_accs1 = pd.read_csv('./test_acc' + str(lrs[0]) + '.csv')

# read the data with lr = 0.001
train_losses2 = pd.read_csv('./train_loss' + str(lrs[1]) + '.csv')
valid_losses2 = pd.read_csv('./valid_loss' + str(lrs[1]) + '.csv')
train_accs2 = pd.read_csv('./test_acc' + str(lrs[1]) + '.csv')

# read the data with lr = 0.01
train_losses3 = pd.read_csv('./train_loss' + str(lrs[2]) + '.csv')
valid_losses3 = pd.read_csv('./valid_loss' + str(lrs[2]) + '.csv')
train_accs3 = pd.read_csv('./test_acc' + str(lrs[2]) + '.csv')
```

```python
# Draw the training loss graph with different learning rates and batch sizes
# 3 figures in line correspond to 3 learning rates
plt.figure(figsize=(12, 7))
plt.subplot(2, 3, 1)
plt.plot(train_losses1['batch_size_4'], label='batch_size_4')
plt.plot(train_losses1['batch_size_32'], label='batch_size_32')
plt.plot(train_losses1['batch_size_64'], label='batch_size_64')
plt.title('Train loss with lr = 0.0001')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0, 3.5)
plt.legend()

plt.subplot(2, 3, 2)
plt.plot(train_losses2['batch_size_4'], label='batch_size_4')
plt.plot(train_losses2['batch_size_32'], label='batch_size_32')
plt.plot(train_losses2['batch_size_64'], label='batch_size_64')
plt.title('Train loss with lr = 0.001')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0, 3.5)
plt.legend()

plt.subplot(2, 3, 3)
plt.plot(train_losses3['batch_size_4'], label='batch_size_4')
plt.plot(train_losses3['batch_size_32'], label='batch_size_32')
plt.plot(train_losses3['batch_size_64'], label='batch_size_64')
plt.title('Train loss with lr = 0.01')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.ylim(0, 3.5)
plt.legend()

# Draw the test accuracy graph with different learning rates and batch sizes
# 3 figures in line correspond to 3 learning rates
plt.subplot(2, 3, 4)
plt.plot(train_accs1['batch_size_4'], label='batch_size_4')
plt.plot(train_accs1['batch_size_32'], label='batch_size_32')
plt.plot(train_accs1['batch_size_64'], label='batch_size_64')
plt.title('Test accuracy with lr = 0.0001')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.legend()

plt.subplot(2, 3, 5)
plt.plot(train_accs2['batch_size_4'], label='batch_size_4')
plt.plot(train_accs2['batch_size_32'], label='batch_size_32')
plt.plot(train_accs2['batch_size_64'], label='batch_size_64')
plt.title('Test accuracy with lr = 0.001')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.legend()

plt.subplot(2, 3, 6)
plt.plot(train_accs3['batch_size_4'], label='batch_size_4')
plt.plot(train_accs3['batch_size_32'], label='batch_size_32')
plt.plot(train_accs3['batch_size_64'], label='batch_size_64')
plt.title('Test accuracy with lr = 0.01')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.legend()

plt.suptitle('VGG19')
plt.tight_layout()
plt.show()
```
